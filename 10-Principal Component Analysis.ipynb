```python
from sklearn.decomposition import PCA
from matplotlib import pyplot as plt
from sklearn.datasets import load_digits
```

```python
data = load_digits()
print(data.data)
```

```python
X_data = data.data
y_data = data.target
```

```python
print(y_data)
```

```python
print(data.data.shape)
```

```python
estimator = PCA(n_components=2)
```

```python
X_pca = estimator.fit_transform(X_data)
```

```python
print(X_pca.shape)
```

```python
colors = ['black', 'blue', 'purple', 'yellow', 'white', 'red', 'lime', 'cyan', 'orange', 'gray']
```

```python
for i in range(len(colors)):
    px = X_pca[:,0][y_data == i]
    py = X_pca[:,1][y_data == i]
    plt.scatter(px,py,c=colors[i])
    plt.legend(data.target_names)
plt.xlabel('First PCA Components')
plt.ylabel('Second PCA Components')
```

```python
from sklearn.model_selection import train_test_split
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
data = fetch_openml('mnist_784')
```

```python
features = data.data
targets = data.target
```

```python
print(features.shape)
```

```python
train_img, test_img, train_lbl, test_lbl = train_test_split(features, targets, test_size=0.15)
```

```python
scaler = StandardScaler()
scaler.fit(train_img)
train_img = scaler.transform(train_img)
test_img = scaler.transform(test_img)
```

```python
pca = PCA(0.95)
```
